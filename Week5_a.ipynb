```python
missing_compensation = df['Compensation'].isna().sum()
print(f"Number of rows with missing calues in 'Compensation' column: {missing_compensation}")
```

```python
def missing_values_summary(df):
  """
  Returns each column name with the number of missing values

  Parameters:
  df (pd.DataFrame): Input dataframe

  Returns:
  pd.Series: Column names with missing value counts
  """

  missing_counts = df.isna().sum()
  return missing_counts[missing_counts >0]

missing_values = missing_values_summary(df)
print("Columns with missing values: ")
print(missing_values)
```

```python
df = df.drop(['Job title context', 'Other currency', 'Salary context'], axis=1)

df.head()
```

```python
rows_before = len(df)
print(f"Rows before: {rows_before}")

df = df.dropna()

rows_after = len(df)
print(f"Rows after: {rows_after}")
print(f"Rows dropped: {rows_before - rows_after}")

df.head()
```

```python
rows_before = len(df)
print(f"Rows before: {rows_before}")

df = df.dropna(thresh=2)

rows_after = len(df)
print(f"Number of rows after: {rows_after}")
print(f"Rows deleted: {rows_before - rows_after}")

df.head()
```

```python
print('--Columns and their number of unique items--')
for column in df.columns:
  n_unique = df[column].nunique()
  print(f'{column} : {n_unique}')
```
