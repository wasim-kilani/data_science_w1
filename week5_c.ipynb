```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.DataFrame(np.random.randn(100, 3))

df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]

print(df)
```

```python
data = {
    'Score': [85, 90, 78, 92, 88, 200, 85, 89, -50, 87]
}
df = pd.DataFrame(data)

print("ORIGINAL DATA: ")
print(df)

df_clean = df[(np.abs(stats.zscore(df)) < 2).all(axis=1)]

print("\nCLEANED DATA (outliers removed): ")
print(df_clean)

print(f"\nSUMMARY: ")
print(f"Original rows: {len(df)}")
print(f"Rows after cleaning: {len(df_clean)}")
print(f"Outliers removed: {len(df) - len(df_clean)}")
```
